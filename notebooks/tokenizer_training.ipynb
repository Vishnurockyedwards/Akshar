# Tokenizer Training
This notebook demonstrates how to train the Hinglish Tokenizer using the merged corpus.

```python
import sys
sys.path.append('..')

from tokenizer.utils import load_and_preprocess_corpus, save_corpus
from tokenizer.tokenizer import HinglishTokenizer

data_dir = '../data'
output_model = '../models/hinglish_tokenizer.json'

print("Loading and preprocessing data...")
corpus = load_and_preprocess_corpus(data_dir)
temp_corpus_path = '../data/merged_corpus_temp.txt'
save_corpus(corpus, temp_corpus_path)
print(f"Total sentences for training: {len(corpus)}")
```

```python
print("Training tokenizer...")
tokenizer = HinglishTokenizer()
tokenizer.train_from_file(temp_corpus_path, vocab_size=130000, min_frequency=2)
tokenizer.save(output_model)
print(f"Model saved to {output_model}")
print(f"Vocab size: {tokenizer.get_vocab_size()}")
```

## Try Custom Sentences
Input your custom sentences below to see how they are tokenized.

```python
from tokenizer.tokenizer import HinglishTokenizer
# Explicitly load the massive trained model from disk
tokenizer = HinglishTokenizer.load('../models/hinglish_tokenizer.json')

custom_sentence = "यहाँ एक यादृच्छिक हिंदी वाक्य है"
encoded = tokenizer.encode(custom_sentence)

# To read Devanagari ByteLevel tokens correctly, we must decode them back to UTF-8!
clean_tokens = [tokenizer.decode([i]) for i in encoded['ids']]

print(f"Original: {custom_sentence}")
print(f"Raw Byte Tokens: {encoded['tokens']}")
print(f"Readable Tokens: {clean_tokens}")
print(f"IDs: {encoded['ids']}")
```
