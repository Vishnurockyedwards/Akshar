# Tokenizer Evaluation
Evaluate the trained Hinglish Tokenizer using basic metrics: vocabulary size, UNK rate, tokens per sentence, and compression ratio.

```python
import sys
import os
import pandas as pd
sys.path.append('..')

from tokenizer.tokenizer import HinglishTokenizer
from tokenizer.utils import load_data

model_path = '../models/hinglish_tokenizer.json'
tokenizer = HinglishTokenizer.load(model_path)
print(f"Loaded Tokenizer with Vocab Size: {tokenizer.get_vocab_size()}")
```

```python
def evaluate_dataset(file_path, tokenizer):
    lines = load_data(file_path)
    if not lines:
        return {}
    
    total_tokens = 0
    total_unks = 0
    total_chars = 0
    
    for line in lines:
        encoded = tokenizer.encode(line)
        tokens = encoded['tokens']
        total_tokens += len(tokens)
        total_unks += sum(1 for t in tokens if t == '[UNK]')
        total_chars += len(line)
        
    return {
        'Language': os.path.basename(file_path),
        'Total Sentences': len(lines),
        'Avg Tokens/Sentence': total_tokens / len(lines),
        'UNK Rate (%)': (total_unks / total_tokens) * 100 if total_tokens else 0,
        'Compression Ratio': total_chars / total_tokens if total_tokens else 0
    }

data_dir = '../data'
results = []
for file in os.listdir(data_dir):
    if file.endswith('.txt') and 'temp' not in file:
        res = evaluate_dataset(os.path.join(data_dir, file), tokenizer)
        results.append(res)

df_results = pd.DataFrame(results)
display(df_results)
```
