# Dataset Analysis
This notebook analyzes the sample datasets.

```python
import os
import pandas as pd
import matplotlib.pyplot as plt

data_dir = '../data'
data_files = [f for f in os.listdir(data_dir) if f.endswith('.txt')]

stats = []
for file in data_files:
    with open(os.path.join(data_dir, file), 'r', encoding='utf-8') as f:
        lines = f.readlines()
        stats.append({
            'Language': file.split('.')[0],
            'Sentences': len(lines),
            'Avg Length': sum(len(line) for line in lines) / len(lines) if lines else 0
        })

df = pd.DataFrame(stats)
display(df)
```

```python
df.plot(x='Language', y='Avg Length', kind='bar', title='Average Sentence Length')
plt.show()
```
